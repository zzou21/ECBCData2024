```python
import nltk, docx, re
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def cosine_similarity(v1, v2):
    dot_product = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)
    return dot_product / (norm_v1 * norm_v2)

vector1 = np.array([11, 1, 3])
vector2 = np.array([14, 4, 5])
vector3 = np.array([7, 10, 13])
vectors = [vector1, vector2, vector3]




similarity = {}
for i in range(len(vectors)):
    for j in range(i+1, len(vectors)):
        similarity[f'vector{i+1}-vector{j+1}'] = cosine_similarity(vectors[i], vectors[j])
print(f"Cosine Similarity: {similarity}")
fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, vector1[0], vector1[1], vector1[2], color='r')
ax.quiver(0, 0, 0, vector2[0], vector2[1], vector2[2], color='b')
ax.quiver(0, 0, 0, vector3[0], vector3[1], vector3[2], color='g')

ax.text(vector1[0], vector1[1], vector1[2], 'V1 - Ciuilitie', color='red', fontsize=18)
ax.text(vector2[0], vector2[1], vector2[2], 'V2 - Gouernment', color='blue', fontsize=18)
ax.text(vector3[0], vector3[1], vector3[2], 'V3 - Susan Constant', color='green', fontsize=18)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([0, 15])
ax.set_ylim([0, 15])
ax.set_zlim([0, 15])
ax.legend()
plt.show()
```
